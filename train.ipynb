```python
import os

import faiss
import numpy as np
import pandas as pd
from sentence_transformers import SentenceTransformer
```

```python
DATA_PATH = "data/cleaned_faq.csv"
MODEL_DIR = "model"
```

```python
INDEX_PATH = os.path.join(MODEL_DIR, "faq_index.faiss")
EMBEDDINGS_PATH = os.path.join(MODEL_DIR, "faq_embeddings.npy")
```

```python
def build_index():
    print("Loading FAQ dataset...")
    df = pd.read_csv(DATA_PATH)
    questions = df["clean_question"].tolist()
    print("Loading embedding model...")
    model = SentenceTransformer("all-MiniLM-L6-v2")
    print("Generating embeddings...")

    embeddings = model.encode(
        questions,
        convert_to_numpy=True,
        show_progress_bar=True
    )

    embeddings = embeddings.astype("float32")
    dimension = embeddings.shape[1]
    index = faiss.IndexFlatL2(dimension)
    index.add(embeddings)
    os.makedirs(MODEL_DIR, exist_ok=True)
    faiss.write_index(index, INDEX_PATH)
    np.save(EMBEDDINGS_PATH, embeddings)

    print(f"Indexed {len(questions)} FAQs")
    print("Training completed successfully!")
```

```python
build_index()
```
